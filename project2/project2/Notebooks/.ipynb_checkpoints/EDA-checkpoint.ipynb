```python
def get_subgroup(dataframe, g_index, g_columns):
    """ check duplicate data """
    g = dataframe.groupby(g_index).agg({col:'nunique' for col in g_columns})
    if g[g > 1].dropna().shape[0] != 0:
        print(" the group has mult same values !")
    return dataframe.groupby(g_index).agg({col:'max' for col in g_columns})

def load_subgroup(filename, index_col=[0]):
    return pd.read_csv(filename, compression='gzip', index_col=index_col)

def save_subgroup(dataframe, g_index, subgroup_name, prefix='raw_'):
    """ save dataframe in csv """
    save_subgroup_filename = "".join([prefix, subgroup_name, ".csv.gz"])
    dataframe.to_csv(save_subgroup_filename, compression='gzip', encoding='UTF-8')
```

# Regression

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import accuracy_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error, r2_score
```

```python
final_players_salary = load_subgroup("final_players_salary.csv.gz")
```

```python
# salary_col = final_players_salary['Salary']
# salary_scaled = (salary_col-salary_col.min())/(salary_col.max()-salary_col.min())
# final_players_salary['salary_scaled']= salary_scaled.copy()
# final_players_salary
```

```python
final_players_salary.columns
features_name = [ 'MP', 'FG', 'FGA', 'P2A', 'P2', 'FT', 'FTA', 'DRB', 'AST',
       'TOV', 'PTS']

X = final_players_salary[features_name]
y = final_players_salary['Salary']
# y = final_players_salary['salary_scaled']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=666)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
```

```python
poly_reg = Pipeline([
    ('poly',PolynomialFeatures(degree=1)),
#     ('poly',PolynomialFeatures(degree=2)),
#     ('std_scaler',StandardScaler()),
    ('std_scaler',StandardScaler()),
    ('lin_reg',LinearRegression())
])
```

```python
poly_reg.fit(X_train,y_train)
poly_reg.score(X_train, y_train)
```

```python
y_train_predict = poly_reg.predict(X_train)
print('Mean squared error: %.2f'% mean_squared_error(y_train, y_train_predict))
print('Coefficient of determination: %.2f'% r2_score(y_train, y_train_predict))
```

```python
poly_reg.score(X_test, y_test)
```

```python
y_test_predict = poly_reg.predict(X_test)
print('Mean squared error: %.2f'  % mean_squared_error(y_test, y_test_predict))
print('Coefficient of determination: %.2f'% r2_score(y_test, y_test_predict))
```

```python
predicted = cross_val_predict(poly_reg, X_train, y_train, cv=10)

fig, ax = plt.subplots()
ax.scatter(y_train, predicted, edgecolors=(0, 0, 0))
ax.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
