# FizzBuzz

```python
import numpy as np
import torch
```

```python
def fizz_buzz_encode(i):
    if i%15 == 0:return 3
    elif i%5==0:return 2
    elif i%3==0:return 1
    else:return 0
    
def fizz_buzz_decode(i,prediction):
    return [str(i),'fizz','buzz','fizzbuzz'][prediction]

def helper(i):
    print(fizz_buzz_decode(i,fizz_buzz_encode(i)))
```

```python
for i in range(1,20):
    helper(i)
```

```python
NUM_DGITS =10
def binary_encode(i,num_digits):
    return np.array([i>>d & 1 for d in range(num_digits)][::-1])


trX = torch.Tensor([binary_encode(i,NUM_DGITS) for i in range(101,2**NUM_DGITS)])
trY = torch.LongTensor([fizz_buzz_encode(i) for i in range(101,2**NUM_DGITS)])

trX.shape
# trY.shape
# binary_encode(31,NUM_DGITS)
```

```python
NUM_HIDDEN = 100
#input [923, 10]
model = torch.nn.Sequential(
    torch.nn.Linear(NUM_DGITS,NUM_HIDDEN), #input [923, 10] * [10, 100] = [923, 100]
    torch.nn.ReLU(),
    torch.nn.Linear(NUM_HIDDEN,4) #input [923, 100] * [100, 4] = [923, 4]
)

loss_fn = torch.nn.CrossEntropyLoss() # 拟合 两个分部相似度
optimizer = torch.optim.SGD(model.parameters(), lr=0.05)
```

```python
batch_size =128
for epoch in range(1000):
    for start in range(0,len(trX),batch_size ):
        end = start +batch_size
        batchX = trX[start:end]
        batchY = trY[start:end]
        
        y_pred = model(batchX)
        loss = loss_fn(y_pred, batchY)
        
        print("epoch",epoch, loss.item())
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
testX = torch.Tensor([binary_encode(i,NUM_DGITS) for i in range(1,101)])
with torch.no_grad():
    testY_pred = model(testX)
# predicts = zip(range(1,101),list(testY_pred.max(1)[1]))
```
